## 1.  Activate your virtual environment

```bash
$ cd Projects/my-project/
$ source venv/bin/activate
$ which python
```

````python
Image("images/activate_venv.png")
````

## 2. Install the IPython Kernel Packages into your virtual environment using Pip-Install

```bash
$ pip install ipykernel
$ pip freeze
```

````python
Image("images/pip_install_ipython_kernel.png")
````

## 3. Register an IPython Kernel using IPyKernel-Install

```bash
$ python -m ipykernel install --user --name my-project --display-name 'My Project'
```
